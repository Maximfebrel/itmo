```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss = pd.read_excel('loss.xlsx')
pred = pd.read_excel('pred.xlsx')
true = pd.read_excel('true.xlsx')
```

```python
# mse
def mean_squared_error(y_pred, y_targ):
    return np.mean(np.square(y_targ - y_pred))

# mse
def mean_absolute_error(y_pred, y_targ):
    return np.mean(abs(y_targ - y_pred))

# rmse
def root_mean_squared_error(y_pred, y_targ):
    return (np.sum((y_targ - y_pred) ** 2) / len(y_targ)) ** 0.5

# r2
def r2_score(y_pred, y_targ):    
    sse = np.sum((y_pred - y_targ) ** 2)
    sst = np.sum((y_pred - np.mean(y_pred)) ** 2)
    return 1 - sse / sst
```

```python
plt.plot(loss[0])
```

```python
true = np.array(true[0])
pred = np.array(pred[0])
```

```python
print('mse: ', mean_squared_error(pred, true))
print('mae: ', mean_absolute_error(pred, true))
print('rmse: ', root_mean_squared_error(pred, true))
print('r2: ', r2_score(pred, true))
```
