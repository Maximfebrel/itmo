# Нейронная сеть

```python
class MLP:
    def __init__(self, architecture):
        self.depth = len(architecture) - 1
        # функция активации
        self.activation_fn = self.sigm
        # производная функции активации по переменной
        self.activation_dfn = self.dsigm
        # функция потерь
        self.loss_fn = self.mse
        # производная функции потерь по переменной
        self.loss_dfn = self.dmse

        # веса и смещение
        self.W = self._init_weights(architecture)
        self.b = self._init_biases(architecture)

        # выход слоя
        self.z = [None] * (self.depth + 1)
        # вход слоя
        self.a = [None] * (self.depth + 1)
        # гриденты весов и смещения
        self.dW = [np.zeros_like(w) for w in self.W]
        self.db = [np.zeros_like(b) for b in self.b]
    
    @staticmethod
    def get_batches(data, batch_size):
        n = len(data)
        get_X = lambda z: z[0:16]
        get_y = lambda z: z[-1]
        for i in range(0, n, batch_size):
            batch = data[i:i+batch_size]
            yield np.array([get_X(b) for b in batch]), np.array([get_y(b) for b in batch])
   
    @staticmethod
    def sigm(x):
        return 1 / (1 + np.exp(-x))
    
    @staticmethod
    def dsigm(x):
        return np.exp(-x) / (1 + np.exp(-x))**2
    
    @staticmethod
    def mse(y_pred, y):
        return np.mean(np.square(y_pred - y))
    
    @staticmethod
    def dmse(y_pred, y):
        n = y.shape[0]
        return (2/n)*(y_pred - y)

    # задаем начальные значения весом при помощи алгоритма рандомизации, при этом задаем ограничения на генерацию
    def _init_weights(self, arch):
        # нейроны на входном слое
        net_in = arch[0]
        # нейроны на выходном слое
        net_out = arch[-1]
        # ограничение
        limit = np.sqrt(6. / (net_in + net_out))
        return [rng.uniform(-limit, limit + 1e-5, size=(arch[i + 1], arch[i])) for i in range(self.depth)]
    
    # задаем начальные значения смещения
    def _init_biases(self, arch):
        return [rng.random((arch[i + 1], 1)) * 2 - 1 for i in range(self.depth)]

    # проходим в прямом направлении по слоям
    def _feedforward(self, X):
        self.a[0] = X.T

        # считаем значения выходов умножением на коэффициенты и добавлением смещения
        for i in range(self.depth):
            self.z[i + 1] = np.matmul(self.W[i], self.a[i]) + self.b[i]

            # умножаем на функцию активации
            self.a[i + 1] = self.activation_fn(self.z[i + 1])
    
    # методом обратного распространения ошибки считаем градиент
    def _backprop(self, y, batch_size=32):
        de_dt = self.loss_dfn(self.a[-1], y)

        for i in range(self.depth - 1, 0, -1):
            if i != self.depth - 1:
                # производная ошибки по переменной для скрытых слоев
                de_dt = self.activation_dfn(self.z[i + 1]) * np.matmul(self.W[i + 1].T, de_dt)
            # производная для весов
            self.dW[i] = np.matmul(de_dt, self.a[i].T)
            # производная для смещения
            self.db[i] = np.sum(de_dt, axis=1, keepdims=True)

    # гралиентный спуск, обновляем веса исходя из скорости обучения и величины градиента
    def _update_params(self, lr=1e-2):
        for i in range(self.depth):
            self.W[i] -= lr * self.dW[i]
            self.b[i] -= lr * self.db[i]
            
    # считаем функцию потерь для предсказанных значений
    def _compute_loss(self, X, y):
        y_pred = self.predict(X).T
        return self.loss_fn(y_pred, y)

    # обучаем модель
    def train(self, X, y, epochs=1, batch_size=32):
        epoch_losses = np.array([])
        dataset = list(zip(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2],
                           X.iloc[:, 3], X.iloc[:, 4], X.iloc[:, 5],
                           X.iloc[:, 6], X.iloc[:, 7], X.iloc[:, 8],
                           X.iloc[:, 9], X.iloc[:, 10], X.iloc[:, 11],
                           X.iloc[:, 12], X.iloc[:, 13], X.iloc[:, 14],
                           X.iloc[:, 15], y.iloc[:, 0]))
        X = list(zip(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2],
                     X.iloc[:, 3], X.iloc[:, 4], X.iloc[:, 5],
                     X.iloc[:, 6], X.iloc[:, 7], X.iloc[:, 8],
                     X.iloc[:, 9], X.iloc[:, 10], X.iloc[:, 11],
                     X.iloc[:, 12], X.iloc[:, 13], X.iloc[:, 14],
                     X.iloc[:, 15]))
        y = np.array(y)
        for i in range(epochs):
            # перемешиваем данные в случайном порядке
            rng.shuffle(dataset)

            # делим генеральную выборку на батчи
            for (X_batch, y_batch) in self.get_batches(dataset, batch_size):
                # прямой метод
                self._feedforward(X_batch)
                # обратный метод
                self._backprop(y_batch)
                # обновление весов
                self._update_params()
            # потери
            l_ = self._compute_loss(X, y)
            epoch_losses = np.append(epoch_losses, l_)
        return epoch_losses

    # выполнение предсказания
    def predict(self, X):
        get_X = lambda z: z[0:16]
        a = np.array([get_X(b) for b in X]).T

        # умножение на веса, добавление смещения, и умножение на функцию активации
        for i in range(self.depth):
            a = np.matmul(self.W[i], a) + self.b[i]
            a = self.activation_fn(a)
        return a
```

# Расчет метрик

Посчитать accuracy, precision, recall, F1-меру
Вывести ROC-кривую

```python
compare_df = pd.DataFrame({'y': list(y_test['class-label']), 'y_pred': y_pred})
```

```python
TP = compare_df.loc[(compare_df['y'] == 1) & (compare_df['y_pred'] == 1)]
FP = compare_df.loc[(compare_df['y'] == 0) & (compare_df['y_pred'] == 1)]
FN = compare_df.loc[(compare_df['y'] == 1) & (compare_df['y_pred'] == 0)]
TN = compare_df.loc[(compare_df['y'] == 0) & (compare_df['y_pred'] == 0)]
```

```python
accuracy = (len(TP) + len(TN)) / len(compare_df)
precession = len(TP) / (len(TP) + len(FP))
recall = len(TP) / (len(TP) + len(FN))
f_1 = 2 * precession * recall / (recall+ precession)
print(f'accuracy:{round(accuracy, 2)}',
     f'precession:{round(precession, 2)}',
     f'recall:{round(recall, 2)}',
     f'f_1-мера:{round(f_1, 2)}',
     sep='\n')
```

```python
_fpr = []
_tpr = []
_def = []

for _ in range(0, 100, 5):
    y_pred_ = []
    for p in pred[0]:
        if p > _ / 100:
            y_pred_.append(1)
        else:
            y_pred_.append(0)
        
    _df = pd.DataFrame({'y': list(y_test['class-label']), 'y_pred': y_pred_})
    _TP = _df.loc[(_df['y'] == 1) & (_df['y_pred'] == 1)]
    _FP = _df.loc[(_df['y'] == 0) & (_df['y_pred'] == 1)]
    _FN = _df.loc[(_df['y'] == 1) & (_df['y_pred'] == 0)]
    _TN = _df.loc[(_df['y'] == 0) & (_df['y_pred'] == 0)]
        
    _fpr.append(len(_FP) / (len(_FP) + len(_TN)))
    _tpr.append(len(_TP) / (len(_TP) + len(_FN)))
    
    _def.append(0.5)

plt.plot(_fpr, _tpr, '-')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC-кривая')
plt.show() 
```
